# Load Data into SQLite Database

This notebook loads the cleaned customer features data into a SQLite database for querying and analysis.

```python
import pandas as pd
import sqlite3

# Load your featured dataset
df = pd.read_csv('c:\\Users\\LENOVO\\Downloads\\customers_features.csv')

# Connect to SQLite database (creates it if not exists)
conn = sqlite3.connect('c:\\Users\\LENOVO\\OneDrive\\Desktop\\sql project\\customer-retention-project\\data\\database\\customers.db')

# Push dataframe into SQLite as a table
df.to_sql('customers', conn, if_exists='replace', index=False)

print("Database created!")
print("Total rows loaded:", pd.read_sql("SELECT COUNT(*) FROM customers", conn).values[0][0])

conn.close()
```
